# Isn't learning one thing? Why do we need multiple tools?

If you only have a hammer, every problem you encounter will seem like a nail. Knowledge comes in various forms, and their uses vary. Therefore it is important to:
1. Be clear on your intention
2. Have quality sources

The tools I use for learning are mindmaps, Zettlekasten, Anki, and flowcharts. Their individual purposes are:
- Mindmaps; making sense of complexity; atomizing knowledge
- Zettlekasten; knowledge creation
- Anki; rote memorization
- Flowchart; deductive reasoning

To create a latticework of mental models, these tools need to be paired with high fidelity knowledge—get to the source.

If you build from trash, you get trash. Build from first principles that have withstood the test of time and scrutiny to generate quality knowledge, and to make better decisions.
